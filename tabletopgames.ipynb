# In a game of guess who, what is the probability of guessing your opponents card on the first try?

- The sample space I am observing is
- One trial is constituted as a game of guess who
- The outcomes I am measuring is how many times the simulation successfully predicts the first card / the number of trials

```python
import random
import matplotlib.pyplot as plt

# Number of character cards in the original Guess Who?
num_cards = 24

# Theoretical probability
theoretical_prob = 1 / num_cards

# Simulation for empirical probability
trials = 1_100_00
correct_guesses = 0

for _ in range(trials):
    opponent_card = random.randint(1, num_cards)
    my_guess = random.randint(1, num_cards)
    if my_guess == opponent_card:
        correct_guesses += 1

empirical_prob = correct_guesses / trials

# Visualization
plt.bar(["Theoretical", "Empirical"], [theoretical_prob, empirical_prob], color="orange")
plt.ylabel("Probability")
plt.title("Probability of Guessing Opponent's Card on First Try (Guess Who?)")
plt.ylim(0, 0.1)
plt.show()

print("Theoretical Probability:", theoretical_prob)
print("Empirical Probability:", empirical_prob)
```

- The chances of guessing your opponents card on the first try is around 0.046 %
- Theoretical Probability: 0.041666666666666664
- Empirical Probability: 0.041772727272727274
