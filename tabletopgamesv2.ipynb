In the game of guess who, what is the probability that a character has no facial hair, given that he is a male character.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Create a dataframe with Guess Who character attributes
characters = pd.DataFrame({
    'Name': ['Alex', 'Alfred', 'Anita', 'Anne', 'Bernard', 'Bill', 'Charles', 
             'Claire', 'David', 'Eric', 'Frans', 'George', 'Herman', 'Joe', 
             'Maria', 'Max', 'Paul', 'Peter', 'Philip', 'Richard', 'Robert', 
             'Sam', 'Susan', 'Tom'],
    'Gender': ['Male', 'Male', 'Female', 'Female', 'Male', 'Male', 'Male', 
               'Female', 'Male', 'Male', 'Male', 'Male', 'Male', 'Male', 
               'Female', 'Male', 'Male', 'Male', 'Male', 'Male', 'Male', 
               'Male', 'Female', 'Male'],
    'Facial_Hair': [False, True, False, False, True, True, True, 
                    False, True, False, False, True, True, True, 
                    False, True, True, False, True, True, False, 
                    True, False, False]
})

# Extract male characters
male_characters = characters[characters['Gender'] == 'Male']
total_males = len(male_characters)
males_no_facial_hair = len(male_characters[male_characters['Facial_Hair'] == False])

# Run simulation for 1000 trials
trials = 1000
probabilities = []

for _ in range(trials):
    # Sample a male character at random
    sample = male_characters.sample(1)
    no_facial_hair = (sample['Facial_Hair'].values[0] == False)
    probabilities.append(int(no_facial_hair))

# Convert results into a DataFrame
results = pd.DataFrame({
    'No_Facial_Hair': probabilities
})

# Plot probability distribution with seaborn
plt.figure(figsize=(8,6))
sns.histplot(results['No_Facial_Hair'], bins=2, discrete=True, stat="probability")
plt.xticks([0,1], ["With Facial Hair", "No Facial Hair"])
plt.title("Probability of No Facial Hair Given Character is Male (1000 trials)")
plt.xlabel("Category")
plt.ylabel("Probability")
plt.show()

# Print probability estimates
prob_estimate = np.mean(probabilities)

print("--- Answer ---")
print(f"1. Analytic calculation: {males_no_facial_hair}/{total_males} = {males_no_facial_hair/total_males:.4f} ({(males_no_facial_hair/total_males)*100:.1f}%).")
print(f"2. Simulation (1000 trials): {sum(probabilities)}/{trials} = {prob_estimate:.4f} ({prob_estimate*100:.1f}%).")
print(f"Therefore, the probability of picking a character with no facial hair, given he is male, is about {(males_no_facial_hair/total_males)*100:.1f}%.")
```

The theoretical probability of randomly choosing a character with no facial hair, given he is male is 31.6% (316/1000)
2. After 1000 Trials the empirical probability was 298/1000, meaning 29.8%
